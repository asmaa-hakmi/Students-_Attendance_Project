```python
# رفع ملف الطلاب
from IPython.display import display
import ipywidgets as widgets
import io
import pandas as pd

uploader_students = widgets.FileUpload(multiple=False)
display(uploader_students)

def upload_students(change):
    global students_df
    uploaded = list(change['new'].values())[0]
    students_df = pd.read_csv(io.BytesIO(uploaded['content']))
    display(students_df.head())

uploader_students.observe(upload_students, names='value')
```

```python
# رفع ملف الحضور
uploader_attendance = widgets.FileUpload(multiple=False)
display(uploader_attendance)

def upload_attendance(change):
    global attendance_df
    uploaded = list(change['new'].values())[0]
    attendance_df = pd.read_csv(io.BytesIO(uploaded['content']))
    display(attendance_df.head())

uploader_attendance.observe(upload_attendance, names='value')
```

```python
# دمج وتحليل البيانات
merged_df = attendance_df.merge(students_df, on='student_id')
merged_df['date'] = pd.to_datetime(merged_df['date'])

absence_summary = (
    merged_df
    .groupby(['cohort', 'branch'])
    .agg(
        total_sessions=('status', 'count'),
        absences=('status', lambda x: (x == 'Absent').sum())
    )
    .reset_index()
)
absence_summary['absence_rate'] = round(100 * absence_summary['absences'] / absence_summary['total_sessions'], 2)
display(absence_summary)
```

```python
# تحليل الغياب حسب أيام الأسبوع
merged_df['day_of_week'] = merged_df['date'].dt.day_name()

day_absence = (
    merged_df[merged_df['status'] == 'Absent']
    .groupby('day_of_week')
    .size()
    .reset_index(name='absence_count')
    .sort_values(by='absence_count', ascending=False)
)
display(day_absence)
```

```python
# رسم بياني لنسبة الغياب حسب الـ cohort
import seaborn as sns
import matplotlib.pyplot as plt

cohort_absence = (
    merged_df
    .groupby(['cohort'])
    .agg(
        total_sessions=('status', 'count'),
        absences=('status', lambda x: (x == 'Absent').sum())
    )
    .reset_index()
)
cohort_absence['absence_rate'] = 100 * cohort_absence['absences'] / cohort_absence['total_sessions']

sns.barplot(x='cohort', y='absence_rate', data=cohort_absence)
plt.title('Absence Rate by Cohort')
plt.ylabel('Absence Rate (%)')
plt.xlabel('Cohort')
plt.show()
```

```python
# Boxplot لتوزيع الغياب حسب الفروع
merged_df['is_absent'] = merged_df['status'].apply(lambda x: 1 if x == 'Absent' else 0)

sns.boxplot(x='branch', y='is_absent', data=merged_df)
plt.title('Absence Distribution by Branch')
plt.ylabel('Absence (1=Absent)')
plt.xlabel('Branch')
plt.show()
```

```python
# تجهيز البيانات
heatmap_data = (
    merged_df[merged_df['status'] == 'Absent']
    .groupby(['branch', 'day_of_week'])
    .size()
    .reset_index(name='absence_count')
    .pivot(index='branch', columns='day_of_week', values='absence_count')
    .fillna(0)
)

# رسم Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap='Reds')
plt.title("Absences Heatmap by Branch and Day of Week")
plt.ylabel("Branch")
plt.xlabel("Day of Week")
plt.show()
```

```python
# حساب عدد الغياب حسب التاريخ
trend = (
    merged_df[merged_df['status'] == 'Absent']
    .groupby('date')
    .size()
    .reset_index(name='absences')
)

# رسم الاتجاه الزمني
plt.figure(figsize=(12, 4))
sns.lineplot(x='date', y='absences', data=trend)
plt.title('Daily Absence Trend')
plt.xlabel('Date')
plt.ylabel('Number of Absences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# تحليل عدد الغياب الكلي لكل فرع
branch_comparison = (
    merged_df[merged_df['status'] == 'Absent']
    .groupby('branch')
    .size()
    .reset_index(name='total_absences')
    .sort_values(by='total_absences', ascending=False)
)
display(branch_comparison)
```

```python
# حساب معدل الغياب لكل طالب
student_absence = (
    merged_df
    .groupby(['student_id', 'name'])
    .agg(
        total_sessions=('status', 'count'),
        absences=('status', lambda x: (x == 'Absent').sum())
    )
    .reset_index()
)
student_absence['absence_rate'] = 100 * student_absence['absences'] / student_absence['total_sessions']

# عرض أكثر 10 طلاب غيابًا فقط
at_risk_top10 = student_absence.sort_values(by='absence_rate', ascending=False).head(10)
display(at_risk_top10[['student_id', 'name', 'absence_rate']])
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x='absence_rate',
    y='name',
    data=at_risk_top10,
    hue='name',
    dodge=False,
    palette='Reds_r',
    legend=False
)
plt.title('Top 10 Students by Absence Rate')
plt.xlabel('Absence Rate (%)')
plt.ylabel('Student Name')
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

```python
import sqlite3

# إنشاء قاعدة بيانات داخل الذاكرة
conn = sqlite3.connect(":memory:")

# رفع البيانات كجداول SQL
students_df.to_sql("students", conn, index=False, if_exists="replace")
attendance_df.to_sql("attendance", conn, index=False, if_exists="replace")
```

```python
#حساب نسبة الغياب الاجمالية
cursor = conn.cursor()
query = """
SELECT
    s.branch,
    COUNT(a.status) AS total_sessions,
    SUM(CASE WHEN a.status = 'Absent' THEN 1 ELSE 0 END) AS total_absences,
    ROUND(100.0 * SUM(CASE WHEN a.status = 'Absent' THEN 1 ELSE 0 END) / COUNT(a.status), 2) AS absence_rate_last_30_days
FROM attendance a
JOIN students s ON a.student_id = s.student_id
GROUP BY s.branch
ORDER BY absence_rate_last_30_days DESC
"""
cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(row)
```

```python
# هذا الاستعلام يعرض اليوم الذي لديه أعلى نسبة حضور.

cursor=conn.cursor()
query="""SELECT CASE strftime('%w',date)
WHEN '0' THEN 'Sunday'
WHEN '1' THEN 'Monday'
WHEN '2' THEN 'Tuesday'
WHEN '3' THEN 'Wednesday'
WHEN '4' THEN 'Thursday'
WHEN '5' THEN 'Friday'
WHEN '6' THEN 'Saturday'
END AS day_of_week,
ROUND(100.0*SUM(CASE WHEN a.status='Present' THEN 1 ELSE 0 END)/COUNT(*),2) AS attendance_rate
FROM attendance a
GROUP BY day_of_week
ORDER BY attendance_rate DESC
LIMIT 1"""
cursor.execute(query)
results=cursor.fetchall()
for row in results:print(row)
```

```python
#حساب معدل الطلاب الذين لم يغيبوا
cursor=conn.cursor()
query="""SELECT s.student_id, s.name
FROM students s
LEFT JOIN attendance a ON s.student_id=a.student_id
GROUP BY s.student_id, s.name
HAVING SUM(CASE WHEN a.status='Absent' THEN 1 ELSE 0 END)=0 OR SUM(CASE WHEN a.status IS NULL THEN 1 ELSE 0 END)=0"""
cursor.execute(query)
results=cursor.fetchall()
for row in results:
    print(row)
```

```python
#: حساب إحصائيات الحضور حسب المجموعة
cursor=conn.cursor()
query="""SELECT CASE WHEN cohort=(SELECT MAX(cohort) FROM students) THEN 'New' ELSE 'Old' END AS cohort_type, COUNT(a.status) AS total_sessions, SUM(CASE WHEN a.status='Absent' THEN 1 ELSE 0 END) AS absences, ROUND(100.0*SUM(CASE WHEN a.status='Absent' THEN 1 ELSE 0 END)/COUNT(a.status), 2) AS absence_rate FROM attendance a JOIN students s ON a.student_id=s.student_id GROUP BY cohort_type"""
cursor.execute(query)
results=cursor.fetchall()
for row in results:print(row)
```

```python
# : حساب متوسط الجلسات لكل مجموعة
cursor=conn.cursor()
query="""SELECT s.branch, ROUND(AVG(session_counts.count_sessions), 2) AS avg_sessions_per_student FROM (SELECT student_id, COUNT(*) AS count_sessions FROM attendance GROUP BY student_id) AS session_counts JOIN students s ON session_counts.student_id=s.student_id GROUP BY s.branch ORDER BY avg_sessions_per_student DESC"""
cursor.execute(query)
results=cursor.fetchall()
for row in results:print(row)
```
